```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/USArrests.csv")
```

```python
data.columns = ["State", "Murder", "Assault", "UrbanPop", "Rape"]
```

```python
crimes = ["Murder", "Assault", "UrbanPop", "Rape"]
```

```python
## Número médio de cada um dos crimes
```

```python
data[crimes].mean()
```

```python
## Mediana de cada um dos crimes
```

```python
data[crimes].median()
```

```python
## Quartis .25, .5, .75 e 1 de cada crime
```

```python
data.quantile([.25, .5, .75, 1])
```

```python
## Mínimos de cada crime
```

```python
data[crimes].min()
```

```python
## Máximos de cada crime
```

```python
data[crimes].max()
```

```python
plt.figure(figsize=(16,12))
plt.bar(data["State"], data["Murder"])
plt.xticks(rotation=90)
plt.show()
```

```python
## Correlação entre os crimes e a população urbana
```

```python
data[["Murder", "Assault", "Rape", "UrbanPop"]].corr()
```

```python
## Maior ocorrência de assassinatos
```

```python
print(data[["State", "Murder"]].max())
print(data[["State", "Assault"]].max())
print(data[["State", "Rape"]].max())
```
