```python
df = pd.read_csv("data/imdb.csv", index_col = 0)
```

```python
positive_budget = df[df["orcamento"] > 0]
positive_budget
```

```python
## Dispersão das notas pelos orçamentos dos filmes
```

```python
positive_budget = df[df["orcamento"] > 0]
```

```python
plt.figure(figsize=(16,12))
plt.scatter(positive_budget["orcamento"] / 1000000, positive_budget["nota_imdb"])
plt.ylabel("Orçamento (em milhões)")
plt.ylabel("Nota IMDB")
plt.show()
```

```python
## Orçamento médio dos filmes ao longo dos anos
```

```python
positive_budget.groupby(["ano"]).mean()
```

```python
average_budget_by_year = positive_budget[positive_budget["ano"] > 0].groupby(["ano"]).mean()
plt.figure(figsize=(16,12))
plt.bar(average_budget_by_year.index, average_budget_by_year["orcamento"] / 1000000)
plt.xlabel("Ano")
plt.ylabel("Orçamento médio (em milhões)")
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.plot(average_budget_by_year.index, average_budget_by_year["orcamento"] / 1000000)
plt.xlabel("Ano")
plt.ylabel("Orçamento médio (em milhões)")
plt.show()
```

```python
positive_budget[["ator_1", "titulo", "orcamento"]]
positive_budget[["ator_1", "orcamento"]][positive_budget["ano"] > 0].groupby(["ator_1"]).agg(["count", "sum"])
```

```python
data = positive_budget[["ator_1", "orcamento"]][positive_budget["ano"] > 0].groupby(["ator_1"]).agg(["count", "sum"]).reset_index()
budgets = data["orcamento"]
first_five = budgets.sort_values(by="count", ascending=False)[:5]
first_five_actors = data.filter(items=first_five.index, axis=0)
first_five_actors
```

```python
actors_data = positive_budget.copy()
```

```python
# actors_data[actors_data["ator_1"] == "Denzel Washington"]
```

```python
plt.figure(figsize=(16,12))
actors_data["orcamento_em_milhoes"] = actors_data["orcamento"] / 1000000
box_plot_data = []
for actor in first_five_actors["ator_1"].values:
    box_plot_data.append(actors_data[actors_data["ator_1"] == actor]["orcamento_em_milhoes"].quantile([.25, .5, .75, 1]))
plt.boxplot(box_plot_data, patch_artist=True, boxprops=dict(facecolor='b'))
plt.xticks([1, 2, 3, 4, 5], first_five_actors["ator_1"].values)
plt.show()
```

```python
bigger_than_8 = positive_budget[positive_budget["ano"] > 0].copy()
bigger_than_8["nota_maior_que_8"] = bigger_than_8["nota_imdb"] > 8
bigger_than_8_by_year = bigger_than_8[bigger_than_8["nota_maior_que_8"] == True].groupby("ano").count()
```

```python
bigger_than_8_by_year.index
```

```python
plt.figure(figsize=(16,12))

plt.bar(bigger_than_8_by_year.index, bigger_than_8_by_year["titulo"])
plt.ylabel("Quantidade de filmes")
plt.xlabel("Ano")
plt.title("Quantidade de filmes por ano com nota superior a 8")
plt.show()
```

```python
plt.figure(figsize=(16,12))

plt.pie(bigger_than_8_by_year["titulo"], labels=bigger_than_8_by_year.index, startangle=90)
plt.title("Quantidade de filmes por ano com nota superior a 8")
plt.axis("equal")
plt.show()
```
